```python
file_name="/Users/apple/Desktop/IGem/后端/compound_emzyme.csv"
f=open(file_name,"r")
list=[]
n=0
for line in f.readlines():
    name=line.split(",,")[0]
    if name in list:
        pass
    else:
        list.append(name)
        n+=1
list=list[1::]

n=0
for cpd in list:
    web_raw="https://www.genome.jp/dbget-bin/www_bget?cpd:"
    web=web_raw+cpd
    n+=1
    print(n,':',web,'   ',cpd)
    
print(n)

```

```python
import urllib.request
import urllib
import re


web_raw="https://www.genome.jp/dbget-bin/www_bget?cpd:"
compound=[]
for cpd in list:
    web=web_raw+cpd
    request=urllib.request.Request(web)
    response=urllib.request.urlopen(request)
    content=response.read().decode('utf-8')
    pattern=re.compile('<div style="width:555px;overflow-x:auto;overflow-y:hidden"><div style="width:555px;overflow-x:auto;overflow-y:hidden">(.*?)<br>',re.S)
    items=re.findall(pattern,content)
    print(cpd," : ",items[0])
    compound.append(items[0])
```
